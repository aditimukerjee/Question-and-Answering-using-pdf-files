# Preprocessing of documents

```python
# This is a default usage of the PreProcessor.
# Here, it performs cleaning of consecutive whitespaces
# and splits a single large document into smaller documents.
# Each document is up to 1000 words long and document breaks cannot fall in the middle of sentences
preprocessor = PreProcessor(
    clean_empty_lines=True,
    clean_whitespace=True,
    clean_header_footer=False,
    split_by="word",
    split_length=100,
    split_respect_sentence_boundary=True
)
```

```python
#importing the dependencies
import nltk
nltk.download('punkt')
```

```python
dict1 = preprocessor.process([doc_pdf])
print(f"n_docs_input: 1\nn_docs_output: {len(dict1)}")
dict1 [:2]
```

```python
print("Num Words in first content: ", len(dict1[0]['content'].split(" ")))
```

## Initalize Retriever, Reader,  & Pipeline

### Retriever

Retrievers help narrowing down the scope for the Reader to smaller units of text where a given question could be answered.

**Here:** We use `TfidfRetriever` in InMemory Document store for simple prototyping and debugging


```python
from haystack.utils import launch_milvus
from haystack.document_stores import MilvusDocumentStore

launch_milvus()
document_store = MilvusDocumentStore()
```

```python
# Now, let's write the dicts containing documents to the database.
document_store.write_documents(dict1)
```

```python
# An in-memory TfidfRetriever based on Pandas dataframes for building quick-prototypes with SQLite document store.
from haystack.nodes import TfidfRetriever, DensePassageRetriever
```

```python
from haystack.nodes import DensePassageRetriever #importing the dense passage retriever from the 
retriever = DensePassageRetriever(document_store=document_store,
                                  query_embedding_model="facebook/dpr-question_encoder-single-nq-base",
                                  passage_embedding_model="facebook/dpr-ctx_encoder-single-nq-base",
                                  max_seq_len_query=64,
                                  max_seq_len_passage=256,
                                  batch_size=16,
                                  use_gpu=True,
                                  embed_title=True,
                                  use_fast_tokenizers=True)
```

```python
document_store.update_embeddings(retriever)
```

```python
retriever
```

### Reader

A Reader scans the texts returned by retrievers in detail and extracts the k best answers. 

Haystack currently supports Readers based on the frameworks FARM and Transformers.

**Here:** a medium sized RoBERTa QA model using a Reader based on FARM (https://huggingface.co/deepset/roberta-base-squad2)

 TransformersReader (leveraging the `pipeline` of the Transformers package)


# TransformersReader

```python
# Load a  local model or any of the QA models on Hugging Face's model hub (https://huggingface.co/models)

reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2", use_gpu=True)
```

# Pipeline

With a Haystack Pipeline you can stick together your building blocks to a search pipeline. One of them is the ExtractiveQAPipeline that combines a retriever and a reader to answer our questions. 

```python
from haystack.pipelines import ExtractiveQAPipeline
pipe = ExtractiveQAPipeline(reader, retriever)
```

# Question 1

```python
# You can configure how many candidates the reader and retriever shall return
# The higher top_k_retriever, the better (but also the slower) your answers. 
prediction = pipe.run(query="What is ohm's law?", params={"Retriever": {"top_k": 10}, "Reader": {"top_k": 3}}
)
```

```python
# Now you can either print the object directly...
from pprint import pprint
pprint(prediction)
```

# Question 2

```python
prediction = pipe.run(query="What is hydro power generation?", params={"Reader": {"top_k": 3}})
```

```python
# ...or use a util to simplify the output
# Change `minimum` to `medium` or `all` to raise the level of detail
print_answers(prediction, details="minimum")
```

A Reader scans the texts returned by retrievers in detail and extracts the k best answers. They are based
